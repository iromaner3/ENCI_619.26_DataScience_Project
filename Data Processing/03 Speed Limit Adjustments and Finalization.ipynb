```python
column_to_move = 'VEH_ADT'
position = 17
column = df.pop(column_to_move)
df.insert(position, column_to_move, column)
df.head(1)
```

```python
# Speed Limit adjustment
speed_limit_2019_to_date = {
    'Major Arterial' : 50,
    'Minor Arterial' : 50,
    'Collector' : 40,
    'Local' : 40,
    'Expressway' : 110,
    'Other' : 40
}

speed_limit_before_2019 = {
    'Major Arterial' : 60,
    'Minor Arterial' : 60,
    'Collector' : 50,
    'Local' : 50,
    'Expressway' : 110,
    'Other' : 40
}
```

```python
# Adjust the SPEEDLMT column
def map_speed_limit(row):
    if row['YEAR'] < 2019:
        return speed_limit_before_2019.get(row['ROADCLASS'], 0)
    else:
        return speed_limit_2019_to_date.get(row['ROADCLASS'], 0)
```

```python
df['SPEEDLMT'] = df.apply(map_speed_limit, axis=1)
```
